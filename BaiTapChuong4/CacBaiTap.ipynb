```python
from math import sqrt
print("Chng trnh tnh din tch Tam Gic")
a=float(input("Nhp cnh a>0:"))
b=float(input("Nhp cnh b>0:"))
c=float(input("Nhp cnh c>0:"))
if (a<=0 or b <=0 or c <=0) or (a+b)<=c or (a+c)<=b or b+c<=a:
 print("Tam gic khng hp l")
else:
 cv=a+b+c
 p=cv/2
 dt=sqrt(p*(p-a)*(p-b)*(p-c))
 print("Din tch =",dt)
```

```python
from random import randint
while True:
    somay = randint(1, 101)
    solandoan = 0
    win = False
    while solandoan < 7:
        solandoan += 1
        songuoi = int(input("Nhp s bn on [1...100]: "))
        print("Ln on th", solandoan)
        if somay == songuoi:
            print("Chc mng bn  on ng, s ca my l", somay)
            win = True
            break
        if somay > songuoi:
            print("Bn on sai, S ca my > s bn on")
        else:
            print("Bn on sai, S ca my < s bn on")
    if not win:
        print("Bn  thua, s ca my l", somay)
    hoi = input("Bn c mun chi tip khng (c/k)? ")
    if hoi=="k":
        break
    print("Cm n bn  chi Game!")
```

```python
def BMI (height, weight):
    return weight / (height **2)
def PhanLoai(bmi):
    if bmi < 18.5:
        return "Gy"
    elif bmi <= 24.9:
        return "Bnh thng"
    elif bmi <= 29.9:
        return "Tha cn"
    elif bmi <= 34.9:
        return "Bo ph cp  1"
    elif bmi <= 39.9:
        return "Bo ph cp  2"
    else:
        return "Bo ph cp  3"
def NguyCoBenh(bmi):
    if bmi < 18.5:
        return "Nguy c bnh thp"
    elif bmi <= 24.9:
        return "Nguy c bnh trung bnh"
    elif bmi <= 29.9:
        return "Nguy c bnh cao"
    elif bmi <= 34.9:
        return "Nguy c bnh cao"
    elif bmi <= 39.9:
        return "Nguy c bnh rt cao"
    else:
        return "Nguy c bnh nguy him"
print("Chng trnh tnh ch s BMI")
height = float(input("Nhp chiu cao (mt): "))
weight = float(input("Nhp cn nng (kg): "))
bmi = BMI(height, weight)
print("Ch s BMI ca bn l: ", BMI(height, weight))
print("Phn loi: ", PhanLoai(bmi))
print("Nguy c bnh ca Thm! l: ", NguyCoBenh(bmi))
```

```python
def fibonacci(n):
     if n<=2 :
        return 1
     return fibonacci(n-1)+fibonacci(n-2)
def listfibo(n):
 for i in range(1,n+1):
     print(fibonacci(i),end='\t')
print(fibonacci(9))
listfibo(9)
```

Hm randrange(0, 100) trong Python s sinh s nguyn ngu nhin trong khong [0, 100), tc l:
- Bt u t 0
- Ln nht l 99
- Ch nhn gi tr nguyn (int), khng bao gm s thc

==> Cc gi tr c th xut hin khi chy randrange(0, 100) l: 34, 0, 99

```python
import random

# Danh sch gi tr cn kim tra
values = [4.5, 34, -1, 100, 0, 99]

print("Kim tra cc gi tr c th xut hin trong randrange(0, 100):\n")
for v in values:
    if isinstance(v, int) and 0 <= v < 100:
        print(f"{v}:  C th xut hin")
    else:
        print(f"{v}:  Khng th xut hin")
"""Hm randrange(0, 100) trong Python s sinh s nguyn ngu nhin trong khong [0, 100), tc l:
Bt u t 0
Ln nht l 99
Ch nhn gi tr nguyn (int), khng bao gm s thc"""
```

```python
import math

"""
Chng trnh tnh  di on thng AB
Nhp ta  im A(xA, yA) v B(xB, yB).
Cng thc: d = sqrt((xB - xA)^2 + (yB - yA)^2)
"""

# Nhp ta  im A v B
xA = float(input("Nhp honh  xA: "))
yA = float(input("Nhp tung  yA: "))
xB = float(input("Nhp honh  xB: "))
yB = float(input("Nhp tung  yB: "))

# Tnh  di on AB
d = math.sqrt((xB - xA)**2 + (yB - yA)**2)

# Xut kt qu
print(f" di on AB = {d}")
```

```python
import math

print("Chng trnh tnh logx (logarit c s a ca x)")
x = float(input("Nhp x (>0): "))
a = float(input("Nhp a (>0, a  1): "))

if x > 0 and a > 0 and a != 1:
    logax = math.log(x) / math.log(a)
    print(f"logx = {logax}")
else:
    print("Gi tr nhp khng hp l (x > 0, a > 0, a  1)")
```
