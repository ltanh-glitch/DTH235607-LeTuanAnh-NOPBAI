```python
def CheckDoiXung(s):
    flag = True
    for i in range(len(s)):
        if s[i] != s[len(s) - i - 1]:
            flag = False
            break
    return flag
def main():
    s = input("Nhập vào 1 chuỗi: ")
    if CheckDoiXung(s):
        print("Chuỗi đối xứng")
    else:
        print("Chuỗi không đối xứng")
while True:
    main()
    s = input("Bạn có muốn tiếp tục không (c/k): ")
    if s == 'k':
        break
print("Chương trình kết thúc")
```

## Một số hàm quan trọng trong xử lý chuỗi của Python

- `len(s)`: Trả về độ dài chuỗi `s`.
- `s.lower()`: Chuyển toàn bộ chuỗi sang chữ thường.
- `s.upper()`: Chuyển toàn bộ chuỗi sang chữ hoa.
- `s.capitalize()`: Viết hoa chữ cái đầu tiên của chuỗi.
- `s.title()`: Viết hoa chữ cái đầu của mỗi từ trong chuỗi.
- `s.strip()`: Loại bỏ khoảng trắng ở đầu và cuối chuỗi.
- `s.lstrip()`, `s.rstrip()`: Loại bỏ khoảng trắng bên trái/phải chuỗi.
- `s.replace(old, new)`: Thay thế tất cả các chuỗi con `old` bằng `new` trong chuỗi.
- `s.split(sep)`: Tách chuỗi thành danh sách các phần tử dựa trên ký tự phân tách `sep`.
- `sep.join(list)`: Nối các phần tử trong list thành chuỗi, ngăn cách bởi `sep`.
- `s.find(sub)`: Trả về vị trí xuất hiện đầu tiên của chuỗi con `sub` trong chuỗi, trả về -1 nếu không tìm thấy.
- `s.count(sub)`: Đếm số lần xuất hiện của chuỗi con `sub` trong chuỗi.
- `s.startswith(prefix)`, `s.endswith(suffix)`: Kiểm tra chuỗi có bắt đầu/kết thúc bằng chuỗi con chỉ định không.

```python
### Chương trình đếm các loại ký tự trong chuỗi

s = input("Nhập vào một chuỗi: ")
so_hoa = so_thuong = so_so = so_db = so_space = so_nguyenam = so_phuam = 0
nguyen_am = 'aeiouAEIOUáàảãạăắằẳẵặâấầẩẫậéèẻẽẹêếềểễệíìỉĩịóòỏõọôốồổỗộơớờởỡợúùủũụưứừửữựýỳỷỹỵÁÀẢÃẠĂẮẰẲẴẶÂẤẦẨẪẬÉÈẺẼẸÊẾỀỂỄỆÍÌỈĨỊÓÒỎÕỌÔỐỒỔỖỘƠỚỜỞỠỢÚÙỦŨỤƯỨỪỬỮỰÝỲỶỸỴ'
for c in s:
    if c.isupper():
        so_hoa += 1
    elif c.islower():
        so_thuong += 1
    elif c.isdigit():
        so_so += 1
    elif c.isspace():
        so_space += 1
    elif not c.isalnum():
        so_db += 1
    # Đếm nguyên âm và phụ âm (chỉ xét ký tự chữ cái)
    if c.isalpha():
        if c in nguyen_am:
            so_nguyenam += 1
        else:
            so_phuam += 1
print("Số chữ IN HOA:", so_hoa)
print("Số chữ in thường:", so_thuong)
print("Số chữ số:", so_so)
print("Số ký tự đặc biệt:", so_db)
print("Số khoảng trắng:", so_space)
print("Số nguyên âm:", so_nguyenam)
print("Số phụ âm:", so_phuam)
```

```python
import re

def NegativeNumberInStrings(s):
    # Tìm tất cả số nguyên âm trong chuỗi
    numbers = re.findall(r'-(?:\d)+', s)
    return [int(num) for num in numbers]

# Nhập chuỗi từ bàn phím
chuoi = input("Nhập vào chuỗi bất kỳ: ")
kq = NegativeNumberInStrings(chuoi)
print("Các số nguyên âm trong chuỗi là:", kq)
```

```python
def ToiUuChuoiDanhTu(s):
    # Xóa khoảng trắng đầu cuối, tách các từ, loại bỏ khoảng trắng thừa
    arr = s.strip().split()
    # Viết hoa chữ cái đầu mỗi từ, các ký tự còn lại viết thường
    arr = [word.capitalize() for word in arr]
    return ' '.join(arr)

# Nhập chuỗi từ bàn phím
chuoi = input("Nhập vào chuỗi danh từ: ")
kq = ToiUuChuoiDanhTu(chuoi)
print("Chuỗi tối ưu:", kq)
```

```python
def LayTenBaiHatDuoi(path):
    # Lấy tên file có đuôi
    return path.strip().split("\\")[-1]

def LayTenBaiHatKhongDuoi(path):
    # Lấy tên file không đuôi
    ten_file = path.strip().split("\\")[-1]
    return ten_file.rsplit('.', 1)[0]

# Nhập đường dẫn từ bàn phím
path = input("Nhập đường dẫn file nhạc: ")
print("Tên bài hát (có đuôi):", LayTenBaiHatDuoi(path))
print("Tên bài hát (không đuôi):", LayTenBaiHatKhongDuoi(path))
```
