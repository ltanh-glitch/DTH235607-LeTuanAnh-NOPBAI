```python
from random import randrange

def nhap_so_nguyen(thong_bao):
    while True:
        try:
            return int(input(thong_bao))
        except ValueError:
            print("Vui lòng nhập đúng số nguyên!")

print("Chương trình xử lý List")
n = nhap_so_nguyen("Nhập số phần tử của List: ")
lst = [randrange(-100, 100) for _ in range(n)]
print("List ngẫu nhiên:", lst)
value = nhap_so_nguyen("Mời bạn thêm số mới vào List: ")
lst.append(value)
print("List sau khi thêm số mới:", lst)
k = nhap_so_nguyen("Bạn muốn đếm số nào trong List?: ")
count = lst.count(k)
print(f"Số {k} xuất hiện {count} lần trong List")
def CheckPrime(n):
    d = 0
    for i in range(1, abs(n) + 1):
        if n % i == 0:
            d += 1
    return d == 2
count_nt = 0
tong_nt = 0
for x in lst:
    if CheckPrime(x):
        count_nt += 1
        tong_nt += x
print(f"Có {count_nt} số nguyên tố trong List")
print(f"Tổng các số nguyên tố trong List là: {tong_nt}")
lst.sort()
print("List sau khi sắp xếp:", lst)
lst.clear()
print("List sau khi xóa:", lst)
# Lưu ý: Nếu bạn chạy lại ô này nhiều lần liên tiếp trên Jupyter, Jupyter sẽ in ra kết quả của mỗi lần chạy. Để tránh lặp, chỉ chạy 1 lần hoặc clear output trước khi chạy lại.
```

```python
from random import randrange

lst = []
n = int(input("Nhập số phần tử của List: "))
for i in range(n):
    lst.append(randrange(0, 100))
print("List sau khi tạo ngẫu nhiên là :", lst);

x = int(input("Mời bạn thêm số mới vào List: "))
lst.append(x)  
print("List sau khi thêm số mới là :", lst);

k = int(input("Mời bạn nhập số để xóa trong List?: "))
while lst.count(k) > 0:
    lst.remove(k)
print("List sau khi xóa số", k, "là:", lst);

def CheckDoiXung(lst):
    for i in range(len(lst)):
        if lst[i] != lst[len(lst) - 1 - i]:
            return False
    return True

kt = CheckDoiXung(lst)
if kt == True:
    print("List đối xứng");
else:
    print("List không đối xứng");

# Thêm dòng này để Jupyter không hiển thị gì thêm
None
```

```python
from random import randrange

def TaoMaTran(m, n):
    D = []
    for i in range(m):
        row = []
        for j in range(n):
            row.append(randrange(100))
        D.append(row)
    return D

def XuatMaTran(D):
    for row in D:
        for element in row:
            print(element, end='\t')
        print()

def LayDong(D, r):
    if 0 <= r < len(D):
        return D[r]
    else:
        print("Dòng không hợp lệ!")
        return []

def XuatList1Chieu(R):
    for element in R:
        print(element, end='\t')
    print()

def LayCot(D, c):
    if len(D) == 0 or c < 0 or c >= len(D[0]):
        print("Cột không hợp lệ!")
        return []
    C = []
    for i in range(len(D)):
        C.append(D[i][c])
    return C

def MAX(D):
    if not D or not D[0]:
        return None
    max_val = D[0][0]
    for i in range(len(D)):
        for j in range(len(D[i])):
            if max_val < D[i][j]:
                max_val = D[i][j]
    return max_val

m = int(input("Nhập số dòng: "))
n = int(input("Nhập số cột: "))
D = TaoMaTran(m, n)
print("Ma trận vừa tạo:")
XuatMaTran(D)
r = int(input("Mời bạn nhập dòng muốn xuất (bắt đầu từ 0): "))
print(f"Dòng {r}:")
XuatList1Chieu(LayDong(D, r))
c = int(input("Mời bạn nhập cột muốn xuất (bắt đầu từ 0): "))
print(f"Cột {c}:")
XuatList1Chieu(LayCot(D, c))
max_val = MAX(D)
print("Số lớn nhất trong ma trận =", max_val)
```
