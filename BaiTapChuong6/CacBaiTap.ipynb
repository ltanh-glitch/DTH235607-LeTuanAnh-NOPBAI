```python
from random import randrange

def TaoMaTran(m, n):
    D = []
    for i in range(m):
        row = []
        for j in range(n):
            row.append(randrange(100))
        D.append(row)
    return D

def XuatMaTran(D):
    for row in D:
        for element in row:
            print(element, end='\t')
        print()

def LayDong(D, r):
    # r: 1-based index
    if 1 <= r <= len(D):
        return D[r-1]
    else:
        print("Dòng không hợp lệ!")
        return []

def XuatList1Chieu(R):
    for element in R:
        print(element, end='\t')
    print()

def LayCot(D, c):
    # c: 1-based index
    if len(D) == 0 or c < 1 or c > len(D[0]):
        print("Cột không hợp lệ!")
        return []
    C = []
    for i in range(len(D)):
        C.append(D[i][c-1])
    return C

def MAX(D):
    if not D or not D[0]:
        return None
    max_val = D[0][0]
    for i in range(len(D)):
        for j in range(len(D[i])):
            if max_val < D[i][j]:
                max_val = D[i][j]
    return max_val

m = int(input("Nhập số dòng: "))
n = int(input("Nhập số cột: "))
D = TaoMaTran(m, n)
print("Ma trận vừa tạo:")
XuatMaTran(D)
r = int(input(f"Mời bạn nhập dòng muốn xuất (bắt đầu từ 1): "))
print(f"Dòng {r}:")
XuatList1Chieu(LayDong(D, r))
c = int(input(f"Mời bạn nhập cột muốn xuất (bắt đầu từ 1): "))
print(f"Cột {c}:")
XuatList1Chieu(LayCot(D, c))
max_val = MAX(D)
print("Số lớn nhất trong ma trận =", max_val)
```

```python
# Câu 7: Nhập vào một dãy các số theo thứ tự tăng dần, nhập sai thì yêu cầu nhập lại
def nhap_so_nguyen(thong_bao):
    while True:
        try:
            return int(input(thong_bao))
        except ValueError:
            print("Vui lòng nhập đúng số nguyên!")

n = nhap_so_nguyen("Nhập số lượng phần tử: ")
lst = []
for i in range(n):
    while True:
        x = nhap_so_nguyen(f"Nhập số thứ {i+1}: ")
        if i == 0 or x > lst[-1]:
            lst.append(x)
            break
        else:
            print("Số vừa nhập phải lớn hơn số trước đó! Vui lòng nhập lại.")
print("Dãy số tăng dần vừa nhập:", lst)
```

```python
# Câu 8: Nhập vào một dãy n số thực, sắp xếp giảm dần và xuất ra dãy số sau khi sắp xếp
def nhap_so_thuc(thong_bao):
    while True:
        try:
            return float(input(thong_bao))
        except ValueError:
            print("Vui lòng nhập đúng số thực!")

n = int(input("Nhập số lượng phần tử n: "))
M = []
for i in range(n):
    x = nhap_so_thuc(f"Nhập số thực M[{i}]: ")
    M.append(x)
M.sort(reverse=True)
print("Dãy số sau khi sắp xếp giảm dần:", M)
```

```python
# Câu 9: Xử lý mảng số tự nhiên theo yêu cầu
def nhap_so_nguyen(thong_bao):
    while True:
        try:
            return int(input(thong_bao))
        except ValueError:
            print("Vui lòng nhập đúng số nguyên!")

def CheckPrime(n):
    if n < 2: return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

n = nhap_so_nguyen("Nhập số lượng phần tử: ")
M = []
for i in range(n):
    x = nhap_so_nguyen(f"Nhập số M[{i}]: ")
    M.append(x)

le = [x for x in M if x % 2 == 1]
chan = [x for x in M if x % 2 == 0]
nt = [x for x in M if CheckPrime(x)]
khong_nt = [x for x in M if not CheckPrime(x)]

print("Dòng 1: Các số lẻ:", le, f"-> Tổng cộng {len(le)} số lẻ")
print("Dòng 2: Các số chẵn:", chan, f"-> Tổng cộng {len(chan)} số chẵn")
print("Dòng 3: Các số nguyên tố:", nt)
print("Dòng 4: Các số không phải là số nguyên tố:", khong_nt)
```

```python
# Câu 10: Xử lý Ma trận: nhập 2 ma trận, cộng và hoán vị
def nhap_ma_tran(m, n, ten):
    print(f"Nhập ma trận {ten} ({m}x{n}):")
    mat = []
    for i in range(m):
        row = []
        for j in range(n):
            x = float(input(f"{ten}[{i+1}][{j+1}]: "))
            row.append(x)
        mat.append(row)
    return mat

def xuat_ma_tran(mat):
    for row in mat:
        print("\t".join(str(x) for x in row))

def cong_ma_tran(A, B):
    m, n = len(A), len(A[0])
    C = []
    for i in range(m):
        row = []
        for j in range(n):
            row.append(A[i][j] + B[i][j])
        C.append(row)
    return C

def hoan_vi_ma_tran(mat):
    m, n = len(mat), len(mat[0])
    trans = []
    for j in range(n):
        row = []
        for i in range(m):
            row.append(mat[i][j])
        trans.append(row)
    return trans

m = int(input("Nhập số dòng của ma trận: "))
n = int(input("Nhập số cột của ma trận: "))
A = nhap_ma_tran(m, n, "A")
B = nhap_ma_tran(m, n, "B")
print("Ma trận A:")
xuat_ma_tran(A)
print("Ma trận B:")
xuat_ma_tran(B)
C = cong_ma_tran(A, B)
print("Tổng 2 ma trận A + B:")
xuat_ma_tran(C)
print("Ma trận hoán vị của A:")
xuat_ma_tran(hoan_vi_ma_tran(A))
print("Ma trận hoán vị của B:")
xuat_ma_tran(hoan_vi_ma_tran(B))
```
